### Задание 1.

Для датафрейма log создать столбец source_type по следующим правилам:

1. Если источник 'traffic_source' равен 'yandex' или 'google', то в 'source_type' ставится 'organic'.
2. Для источников 'paid' и 'email' из России - ставим 'ad'.
3. Для источников 'paid' и 'email' не из России - ставим 'other'.
4. Все остальные варианты берем из 'traffic_source' без изменений.

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')

log.loc[log.traffic_source.isin(['yandex', 'google']), 'source_type'] = 'organic'

log.loc[log.traffic_source.isin(['paid', 'email']) & log.region.isin(['Russia']), 'source_type'] = 'ad'

log.loc[log.traffic_source.isin(['paid', 'email']) & ~log.region.isin(['Russia']), 'source_type'] = 'other'

log.loc[log.source_type.isnull(), 'source_type'] = log['traffic_source']
```

### Задание 3.
Использовать файл с оценками фильмов 'ml-latest-small/ratings.csv'.
Посчитать среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца 'timestamp' для данного значения 'userId'.

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv', sep=',')
```

```python
ratings_count = ratings.groupby('userId').count().reset_index()
ratings_count = ratings_count[['userId', 'rating']] # Берем только нужные столбцы.

age_count = ratings_count[ratings_count['rating'] > 100]
# убираем предупреждение SettingWithCopyWarning.
pd.options.mode.chained_assignment = None
age_count.rename(columns={'rating': 'rating_count'}, inplace=True)
```

```python
ratings_max = ratings.groupby('userId').max().reset_index()
ratings_min = ratings.groupby('userId').min().reset_index()
# берем только нужные столбцы. 
ratings_max = ratings_max[['userId', 'timestamp']]
ratings_min = ratings_min[['userId', 'timestamp']]
# меняем название нужного стольбца в df с max и min значениями.
ratings_max.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)
ratings_min.rename(columns={'timestamp': 'timestamp_min'}, inplace=True)
```

```python
# объединяем все значения в общую таблицу.
average_timestamp = age_count.merge(ratings_max, how='inner').merge(ratings_min, how='inner')
```

```python
def average(row):
    ''' Возвращает разницу между максимальным и минимальным значениями "timestamp". 
    '''
    difference = row['timestamp_max'] - row['timestamp_min']
    return difference
```

```python
average_timestamp['average'] = average_timestamp.apply(average, axis=1)
average_timestamp.head()
```

### Задание 4.
Дана статистика услуг перевозок клиентов компании по типам. Необходимо сформировать две таблицы:
 1. Таблицу с тремя типами выручки для каждого client_id без указания адреса клиента.
 2. Аналогичную таблицу по типам выручки с указанием адреса клиента.

В процессе объединения таблиц данные не должны теряться.

```python
# железнодорожные перевозки.
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
```

```python
# автомобильные перевозки.
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
```

```python
# воздушные перевозки.
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
```

```python
# адреса клиентов.
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
```

```python
# объединяем таблицы с помощью условия 'outer', для сохранения всех строк. 
revenue_table = rzd.merge(auto, how='outer', on='client_id').merge(air, how='outer', on='client_id')

# меняем значения NaN на 0 с помощью фильтра 'isnull'.
revenue_table.loc[revenue_table.rzd_revenue.isnull(), 'rzd_revenue'] = 0
revenue_table.loc[revenue_table.auto_revenue.isnull(), 'auto_revenue'] = 0
revenue_table.loc[revenue_table.air_revenue.isnull(), 'air_revenue'] = 0

revenue_table
```

```python
# Аналогичная таблица, но с добавленными адресами клиентов из df 'client_base'.
revenue_table_with_address = revenue_table.merge(client_base, how='outer', on='client_id')
revenue_table_with_address
```
